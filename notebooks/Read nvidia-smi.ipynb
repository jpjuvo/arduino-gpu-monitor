```python
import subprocess
import pprint
```

```python
sp = subprocess.Popen(['nvidia-smi', '-q'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

out_str = sp.communicate()[0]
if isinstance(out_str[0], bytes):
    out_str = out_str.decode('utf-8')
out_list = out_str.split('\n')

out_dict = {}

for item in out_list:
    try:
        key, val = item.split(':')
        key, val = key.strip(), val.strip()
        out_dict[key] = val
    except:
        pass

pprint.pprint(out_dict)
```

```python
import pynvml

try:
    pynvml.nvmlInit()
except pynvml.NVMLError as error:
    print(error)

try:
    print(pynvml.nvmlDeviceGetCount())
except pynvml.NVMLError as error:
    print(error)

print(pynvml.nvmlDeviceGetCount())# total gpu count = 1
print(pynvml.nvmlSystemGetDriverVersion()) # 396.54

GPU_ID = 0
handle = pynvml.nvmlDeviceGetHandleByIndex(GPU_ID)
print(pynvml.nvmlDeviceGetName(handle)) # GeForce GTX 1060

utilization_rates = pynvml.nvmlDeviceGetUtilizationRates(handle)
print(pynvml.nvmlDeviceGetUtilizationRates(handle))

meminfo = pynvml.nvmlDeviceGetMemoryInfo(handle)
MB_SIZE = 1024*1024
print(meminfo.total/MB_SIZE) # 6078 MB
print(meminfo.used/MB_SIZE)  # 531 MB
print(meminfo.free/MB_SIZE)  # 5546 MB

pynvml.nvmlShutdown()
```
